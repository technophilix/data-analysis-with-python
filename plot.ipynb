```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel('sampledatafoodsales.xlsx', 'FoodSales') #read_csv
```

```python
x=df['Quantity']
y=df['UnitPrice']
```

```python
plt.bar(x,y, label="Sales Data", color='red')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Graph for sales')
#plt.legend()
plt.show()
```
